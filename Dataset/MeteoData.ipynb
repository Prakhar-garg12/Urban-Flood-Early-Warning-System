```python
import requests
import pandas as pd

# Change lat/lon to your city
lat, lon = 23.2599, 77.4126  # Bhopal

url = "https://archive-api.open-meteo.com/v1/archive"
params = {
    "latitude": lat,
    "longitude": lon,
    "start_date": "2010-01-01",
    "end_date": "2026-02-18",
    "daily": [
        "precipitation_sum",
        "temperature_2m_max",
        "temperature_2m_min", 
        "et0_fao_evapotranspiration",
        "rain_sum",
        "precipitation_hours"
    ],
    "timezone": "Asia/Kolkata"
}

response = requests.get(url, params=params)
data = response.json()
df = pd.DataFrame(data["daily"])
df.to_csv("rainfall_data.csv", index=False)
print(df.head())
```

```python
import requests
import pandas as pd

lat, lon = 23.2599, 77.4126  # Bhopal

print("Fetching daily weather variables...")

url = "https://archive-api.open-meteo.com/v1/archive"
params = {
    "latitude": lat,
    "longitude": lon,
    "start_date": "2010-01-01",
    "end_date": "2026-02-18",
    "daily": [
        "temperature_2m_max",
        "temperature_2m_min",
        "temperature_2m_mean",
        "dewpoint_2m_max",
        "dewpoint_2m_min",
        "wind_speed_10m_max",
        "wind_gusts_10m_max",
        "wind_direction_10m_dominant",
        "et0_fao_evapotranspiration",
        "shortwave_radiation_sum"
    ],
    "timezone": "Asia/Kolkata"
}

response = requests.get(url, params=params)
data = response.json()

if "error" in data:
    print("Error:", data["reason"])
else:
    df_weather = pd.DataFrame(data["daily"])
    df_weather["time"] = pd.to_datetime(df_weather["time"])
    df_weather.to_csv("weather_data.csv", index=False)

    print("Shape:", df_weather.shape)
    print("\nColumns:", df_weather.columns.tolist())
    print("\nSample:\n", df_weather.head())
    print("\nMissing values:\n", df_weather.isnull().sum())
```

```python
import requests
import pandas as pd

lat, lon = 23.2599, 77.4126  # Bhopal

print("Fetching hourly soil moisture...")

url = "https://archive-api.open-meteo.com/v1/archive"
params = {
    "latitude": lat,
    "longitude": lon,
    "start_date": "2010-01-01",
    "end_date": "2026-02-18",
    "hourly": [
        "soil_moisture_0_to_7cm",
        "soil_moisture_7_to_28cm",
        "soil_moisture_28_to_100cm"
    ],
    "timezone": "Asia/Kolkata"
}

response = requests.get(url, params=params)
data = response.json()

if "error" in data:
    print("Error:", data["reason"])
else:
    df_hourly = pd.DataFrame(data["hourly"])
    df_hourly["time"] = pd.to_datetime(df_hourly["time"])

    # Resample hourly → daily mean
    df_hourly["date"] = df_hourly["time"].dt.date
    df_soil = df_hourly.groupby("date").mean(numeric_only=True).reset_index()
    df_soil["time"] = pd.to_datetime(df_soil["date"])
    df_soil = df_soil.drop(columns=["date"])

    df_soil.to_csv("soil_moisture_data.csv", index=False)

    print("Shape:", df_soil.shape)
    print("\nColumns:", df_soil.columns.tolist())
    print("\nSample:\n", df_soil.head())
    print("\nMissing values:\n", df_soil.isnull().sum())
```

```python
import pandas as pd
import numpy as np

# ── Load all 4 CSVs ──
print("Loading all datasets...")
df_rain    = pd.read_csv("rainfall_data.csv",        parse_dates=["time"])
df_river   = pd.read_csv("river_discharge_data.csv", parse_dates=["time"])
df_weather = pd.read_csv("weather_data.csv",         parse_dates=["time"])
df_soil    = pd.read_csv("soil_moisture_data.csv",   parse_dates=["time"])

# ── Drop duplicate columns from weather 
# (temperature_2m_max, temperature_2m_min, et0 already exist in rainfall) ──
df_weather = df_weather.drop(columns=[
    "temperature_2m_max",
    "temperature_2m_min",
    "et0_fao_evapotranspiration"
])

print(f"Rainfall shape     : {df_rain.shape}")
print(f"River shape        : {df_river.shape}")
print(f"Weather shape      : {df_weather.shape}")
print(f"Soil moisture shape: {df_soil.shape}")

# ── Merge all 4 on 'time' ──
print("\nMerging all datasets...")
df = df_rain.merge(df_river,   on="time", how="inner")
df = df.merge(df_weather,      on="time", how="inner")
df = df.merge(df_soil,         on="time", how="inner")
df = df.sort_values("time").reset_index(drop=True)
print(f"After merge shape: {df.shape}")

# ── Engineer antecedent rainfall features ──
print("\nEngineering features...")
df["rainfall_3d"]  = df["precipitation_sum"].rolling(window=3).sum()
df["rainfall_7d"]  = df["precipitation_sum"].rolling(window=7).sum()
df["rainfall_14d"] = df["precipitation_sum"].rolling(window=14).sum()
df["rainfall_30d"] = df["precipitation_sum"].rolling(window=30).sum()

# ── Rainfall intensity ──
df["rainfall_intensity"] = df.apply(
    lambda row: row["precipitation_sum"] / row["precipitation_hours"]
    if row["precipitation_hours"] > 0 else 0, axis=1
)

# ── Humidity from dewpoint + temperature ──
df["humidity"] = 100 - 5 * (
    df["temperature_2m_max"] - df["dewpoint_2m_max"]
)
df["humidity"] = df["humidity"].clip(0, 100)

# ── Soil saturation index (avg of top 2 layers) ──
df["soil_saturation"] = (
    df["soil_moisture_0_to_7cm"] +
    df["soil_moisture_7_to_28cm"]
) / 2

# ── Temporal features ──
df["day_of_year"] = df["time"].dt.dayofyear
df["month"]       = df["time"].dt.month

# Cyclical encoding
df["day_sin"]   = np.sin(2 * np.pi * df["day_of_year"] / 365)
df["day_cos"]   = np.cos(2 * np.pi * df["day_of_year"] / 365)
df["month_sin"] = np.sin(2 * np.pi * df["month"] / 12)
df["month_cos"] = np.cos(2 * np.pi * df["month"] / 12)

# ── Monsoon flag ──
df["is_monsoon"] = df["month"].apply(lambda m: 1 if 6 <= m <= 9 else 0)

# ── Days since last significant rainfall ──
df["had_rain"] = (df["precipitation_sum"] > 1).astype(int)
df["days_since_rain"] = df["had_rain"].groupby(
    (df["had_rain"] == 1).cumsum()
).cumcount()
df = df.drop(columns=["had_rain"])

# ── Drop NaN rows from rolling windows ──
df = df.dropna().reset_index(drop=True)

# ── Save final dataset ──
df.to_csv("flood_final_dataset.csv", index=False)

print(f"\nFinal dataset shape : {df.shape}")
print(f"Total features      : {len(df.columns) - 1}  (excluding time)")
print(f"Date range          : {df['time'].min()} → {df['time'].max()}")
print("\nAll columns:")
for i, col in enumerate(df.columns):
    print(f"  {i+1:02d}. {col}")
print("\nFirst 3 rows:")
print(df.head(3))
print(f"\nTotal missing values: {df.isnull().sum().sum()}")
 
```
